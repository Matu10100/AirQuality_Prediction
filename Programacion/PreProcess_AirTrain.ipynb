Creamos una función para calcular el porcentaje de datos nulos en el dataset

```python
def missing_values(df):
    Column_Names = df.columns

    Total_rows = df.shape[0]

    missing_values = df.isnull().sum()

    Percent_missing = (missing_values / Total_rows)*100
    
    result = pd.DataFrame({
        'Column_Name': Column_Names,
        'Missing_Values': missing_values.values,
        'Total_Rows': Total_rows,
        'Percent_Missing': Percent_missing.values
    })

    return result
```

Desde el diccionario de datos del SIATA menciona que valores de calidad mayores a 2.6 no son confiables por lo que se eliminan de mi dataset en aras de mejorar la calidad de mis datos y por ende la predicción

```python
prueba = Air_organized.copy()

# Filtrar todas las columnas de 'calidad_pm25'
calidad_columns = prueba.xs('calidad_pm25', level=0, axis=1)

# Iterar sobre cada código serial en las columnas 'calidad_pm25'
for codigoSerial in calidad_columns:
    # Reemplazar los valores en 'pm25' si 'calidad_pm25' es menor a 1
    prueba.loc[prueba[('calidad_pm25', codigoSerial)] > 1, ('pm25', codigoSerial)] = np.nan
    #prueba.loc[prueba[('calidad_pm25', codigoSerial)] > 1, ('calidad_pm25', codigoSerial)] = np.nan
```

```python
missing_values(prueba)
```

```python
# Crear columnas auxiliares de año, mes, día a partir del índice
prueba['Año'] = prueba.index.year
prueba['Mes'] = prueba.index.month
prueba['Día'] = prueba.index.day

# Calcular la media horaria para cada día específico
# Esto incluye agrupar por año, mes, día  y luego calcular la media
medias_horarias = prueba.groupby(['Año', 'Mes', 'Día']).transform('mean')

# Rellenar valores faltantes usando las medias horarias calculadas
prueba.fillna(medias_horarias, inplace=True)

# Eliminar las columnas auxiliares si no son necesarias
prueba.drop(['Año', 'Mes', 'Día'], axis=1, inplace=True)
```

Para los datos nulos, se tienen 2 alternativas

1. **Forward or backward fill:** Use the value from the previous or next observation to fill null values. This method is especially useful for time series data.

df.fillna(method='ffill', inplace=True, axis=0)

2. **Interpolation:** Estimate missing values based on other values in the dataset. Pandas provides various interpolation methods such as linear, polynomial, or spline interpolation.

df.interpolate(method='linear', inplace=True, axis=0)

Inicialmente se llevara a cabo **Forward or backward fill** pues es más acorde al tipo de datos

```python
# Forward fill null values
intento = prueba['pm25'].copy()
intento.ffill(axis=0, inplace=True)
```

```python
intento.columns.name = None
# Resetear el índice para tener 'Fecha_Hora' como columna
intento.reset_index(inplace=True)
intento.rename(columns={'index': 'Fecha_Hora'}, inplace=True)

# Mapeo de códigos seriales a comunas
mapa_comunas = {12: 10, 79: 16, 80: 8, 83: 16, 84: 14, 85: 10, 86: 4}

# Transformar el dataframe usando melt para desglosar por 'codigoSerial'
df_transformado = intento.melt(id_vars=['Fecha_Hora'], var_name='codigoSerial', value_name='pm_25')

# Crear una columna 'comuna' basada en el código serial
df_transformado['comuna'] = df_transformado['codigoSerial'].map(mapa_comunas)

# Resultado final
print(df_transformado)
```

```python
df_transformado.reset_index(drop=True, inplace=True)  # Esto elimina cualquier índice actual si existe

# Establecer 'Fecha_Hora' como el índice único
df_transformado.set_index('Fecha_Hora', inplace=True)
```
